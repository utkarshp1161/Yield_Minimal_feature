```python
# =============================================================================
# MINIMAL FEATURES IDENTIFICATION SYSTEM - FIXED VERSION
# Compatible with XGBoost 2.0+ and optimized for large datasets
# =============================================================================

print("="*80)
print("MINIMAL FEATURES IDENTIFICATION SYSTEM - v2.0")
print("District √ó Crop √ó Season Specific Analysis")
print("="*80)

# Install packages
import subprocess
import sys

def install(package):
    subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", package])

try:
    import xgboost
    print(f"‚úì XGBoost {xgboost.__version__} installed")
except:
    print("Installing XGBoost...")
    install("xgboost")
    import xgboost

try:
    import shap
    print(f"‚úì SHAP installed")
except:
    print("Installing SHAP...")
    install("shap")

import os
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.preprocessing import StandardScaler, LabelEncoder
import matplotlib.pyplot as plt
import matplotlib.backends.backend_pdf
import seaborn as sns
import joblib
import warnings
from datetime import datetime
warnings.filterwarnings('ignore')

RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)

# Check XGBoost version for API compatibility
xgb_version = tuple(map(int, xgboost.__version__.split('.')[:2]))
print(f"XGBoost version detected: {xgb_version}")

# Upload file
print("\nüìÅ Upload your dataset CSV file:")
from google.colab import files
uploaded = files.upload()
DATA_FILE = list(uploaded.keys())[0]

OUTPUT_DIR = '/content/minimal_features_results'
os.makedirs(OUTPUT_DIR, exist_ok=True)

# =============================================================================
# STEP 1: LOAD DATA
# =============================================================================
print("\n" + "="*80)
print("[1/10] Loading data...")
print("="*80)

df_raw = pd.read_csv(DATA_FILE)
print(f"\n‚úì Loaded: {df_raw.shape[0]:,} rows √ó {df_raw.shape[1]} columns")

df = df_raw.copy()
df.columns = df.columns.str.lower().str.strip().str.replace(' ', '_').str.replace('[^a-z0-9_]', '', regex=True)

# Find key columns
district_col = next((col for col in ['district', 'region'] if col in df.columns), None)
crop_col = next((col for col in ['crop', 'crop_name'] if col in df.columns), None)
season_col = next((col for col in ['season', 'season_name'] if col in df.columns), None)
year_col = next((col for col in ['year', 'crop_year'] if col in df.columns), None)
yield_col = next((col for col in ['yield', 'yield_t_ha'] if col in df.columns), None)

print(f"\n‚úì Key columns:")
print(f"  District: {district_col}")
print(f"  Crop: {crop_col}")
print(f"  Season: {season_col}")
print(f"  Year: {year_col}")
print(f"  Yield: {yield_col}")

print(f"\nüìä Data distribution:")
print(f"  Districts: {df[district_col].nunique():,} unique")
print(f"  Crops: {df[crop_col].nunique()} unique")
print(f"  Seasons: {df[season_col].nunique()} unique")
print(f"  Years: {df[year_col].min()} - {df[year_col].max()}")
print(f"  Yield: {df[yield_col].min():.2f} - {df[yield_col].max():.2f}")

# =============================================================================
# STEP 2: CLEAN DATA
# =============================================================================
print("\n" + "="*80)
print("[2/10] Cleaning data...")
print("="*80)

# Remove empty/constant columns
empty_cols = [col for col in df.columns if df[col].isna().all() or df[col].nunique() == 1]
if empty_cols:
    print(f"Removing {len(empty_cols)} empty/constant columns")
    df = df.drop(columns=empty_cols)

# Remove leakage
leakage_cols = [col for col in df.columns if 'production' in col and col != yield_col]
if leakage_cols:
    print(f"Removing data leakage: {leakage_cols}")
    df = df.drop(columns=leakage_cols)

# Handle missing values
missing = df.isnull().sum()
missing = missing[missing > 0]

if len(missing) > 0:
    print(f"\nHandling {len(missing)} columns with missing values...")
    for col in missing.index:
        if col in [district_col, crop_col, season_col, yield_col]:
            continue

        missing_pct = (missing[col] / len(df)) * 100
        if missing_pct > 70:
            df = df.drop(columns=[col])
            print(f"  {col}: {missing_pct:.1f}% missing - DROPPED")
        elif df[col].dtype in ['float64', 'int64']:
            df[col] = df.groupby([district_col, crop_col, season_col])[col].transform(
                lambda x: x.fillna(x.median())
            )
            df[col].fillna(df[col].median(), inplace=True)
        else:
            df[col].fillna(df[col].mode()[0], inplace=True)

print(f"\n‚úì Cleaned: {df.shape[0]:,} rows √ó {df.shape[1]} columns")

# =============================================================================
# STEP 3: PREPARE FEATURES
# =============================================================================
print("\n" + "="*80)
print("[3/10] Preparing features...")
print("="*80)

# Group identifier
df['group_id'] = df[district_col].astype(str) + '_' + df[crop_col].astype(str) + '_' + df[season_col].astype(str)

# Encode categoricals
encoders = {}
for col in [district_col, crop_col, season_col]:
    le = LabelEncoder()
    df[f'{col}_encoded'] = le.fit_transform(df[col])
    encoders[col] = le

# Get numeric features
exclude_cols = [district_col, crop_col, season_col, year_col, yield_col, 'group_id', 'state']
numeric_features = [col for col in df.columns
                   if col not in exclude_cols
                   and not col.endswith('_encoded')
                   and df[col].dtype in ['int64', 'float64']]

encoded_features = [f'{district_col}_encoded', f'{crop_col}_encoded', f'{season_col}_encoded']
all_features = encoded_features + numeric_features

print(f"‚úì Features: {len(all_features)} total")
print(f"  Categorical (encoded): {len(encoded_features)}")
print(f"  Numeric: {len(numeric_features)}")

# =============================================================================
# STEP 4: TRAIN/TEST SPLIT
# =============================================================================
print("\n" + "="*80)
print("[4/10] Splitting data...")
print("="*80)

X = df[all_features].copy()
y = df[yield_col].copy()

# Stratified split by crop
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_SEED,
    stratify=df[crop_col]
)

X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train, test_size=0.15, random_state=RANDOM_SEED,
    stratify=df.loc[X_train.index, crop_col]
)

# Scale
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)

print(f"‚úì Split: Train={len(X_train):,}, Val={len(X_val):,}, Test={len(X_test):,}")

# =============================================================================
# STEP 5: TRAIN FULL MODEL
# =============================================================================
print("\n" + "="*80)
print("[5/10] Training full model...")
print("="*80)

import xgboost as xgb

# Use callbacks for early stopping (XGBoost 2.0+ compatible)
model_full = xgb.XGBRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=6,
    random_state=RANDOM_SEED,
    n_jobs=-1,
    tree_method='hist'
)

print(f"Training with {len(all_features)} features...")
model_full.fit(X_train_scaled, y_train, verbose=False)

y_pred_val_full = model_full.predict(X_val_scaled)
y_pred_test_full = model_full.predict(X_test_scaled)

r2_val_full = r2_score(y_val, y_pred_val_full)
r2_test_full = r2_score(y_test, y_pred_test_full)
rmse_val_full = np.sqrt(mean_squared_error(y_val, y_pred_val_full))
rmse_test_full = np.sqrt(mean_squared_error(y_test, y_pred_test_full))

print(f"\n{'='*60}")
print(f"FULL MODEL ({len(all_features)} features)")
print(f"{'='*60}")
print(f"  Val R¬≤:   {r2_val_full:.4f} | RMSE: {rmse_val_full:.4f}")
print(f"  Test R¬≤:  {r2_test_full:.4f} | RMSE: {rmse_test_full:.4f}")
print(f"{'='*60}")

# =============================================================================
# STEP 6: FEATURE IMPORTANCE
# =============================================================================
print("\n" + "="*80)
print("[6/10] Computing feature importance...")
print("="*80)

try:
    import shap
    print("Using SHAP (this may take a few minutes for large datasets)...")

    # Use smaller sample for SHAP
    sample_size = min(500, len(X_train_scaled))
    sample_train = np.random.choice(len(X_train_scaled), size=sample_size, replace=False)
    sample_val = np.random.choice(len(X_val_scaled), size=min(200, len(X_val_scaled)), replace=False)

    explainer = shap.Explainer(model_full, X_train_scaled[sample_train])
    shap_values = explainer(X_val_scaled[sample_val])
    importance_values = np.abs(shap_values.values).mean(axis=0)
    method = "SHAP"
    print("‚úì SHAP completed")

except Exception as e:
    print(f"SHAP failed: {e}")
    print("Using XGBoost feature importance...")
    importance_values = model_full.feature_importances_
    method = "XGBoost Gain"

importance_df = pd.DataFrame({
    'feature': all_features,
    'importance': importance_values
}).sort_values('importance', ascending=False).reset_index(drop=True)

importance_df['rank'] = range(1, len(importance_df) + 1)

print(f"\nTop 15 Features ({method}):")
print(importance_df[['rank', 'feature', 'importance']].head(15).to_string(index=False))

# =============================================================================
# STEP 7: ABLATION STUDY
# =============================================================================
print("\n" + "="*80)
print("[7/10] Finding minimal feature set...")
print("="*80)

feature_order = importance_df['feature'].tolist()
target_performance = r2_val_full * 0.95

print(f"Target: {target_performance:.4f} (95% of full model)")
print("\nTesting different feature counts...\n")

ablation_results = []
test_counts = [3, 5, 7, 10, 15, 20, 30, 40, len(all_features)]

for n_features in test_counts:
    if n_features > len(feature_order):
        continue

    current_features = feature_order[:n_features]
    feature_indices = [all_features.index(f) for f in current_features]

    X_train_sub = X_train_scaled[:, feature_indices]
    X_val_sub = X_val_scaled[:, feature_indices]

    model_sub = xgb.XGBRegressor(
        n_estimators=200,
        learning_rate=0.05,
        max_depth=6,
        random_state=RANDOM_SEED,
        n_jobs=-1,
        tree_method='hist'
    )

    model_sub.fit(X_train_sub, y_train, verbose=False)
    y_pred_val_sub = model_sub.predict(X_val_sub)

    r2_val_sub = r2_score(y_val, y_pred_val_sub)
    rmse_val_sub = np.sqrt(mean_squared_error(y_val, y_pred_val_sub))
    performance_pct = (r2_val_sub / r2_val_full) * 100

    ablation_results.append({
        'n_features': n_features,
        'r2': r2_val_sub,
        'rmse': rmse_val_sub,
        'performance_pct': performance_pct
    })

    status = "‚úì TARGET" if r2_val_sub >= target_performance else ""
    print(f"  {n_features:3d} features: R¬≤={r2_val_sub:.4f} ({performance_pct:.1f}%) {status}")

ablation_df = pd.DataFrame(ablation_results)

# Find minimal
minimal_idx = ablation_df[ablation_df['r2'] >= target_performance]['n_features'].min()
minimal_features = feature_order[:minimal_idx]

print(f"\n{'='*60}")
print(f"MINIMAL FEATURE SET: {minimal_idx} features")
print(f"{'='*60}")
print(f"  Performance: {(ablation_df[ablation_df['n_features']==minimal_idx]['r2'].values[0]/r2_val_full*100):.1f}%")
print(f"  Reduction: {100-(minimal_idx/len(all_features)*100):.1f}%")
print(f"{'='*60}")

# =============================================================================
# STEP 8: TRAIN MINIMAL MODEL
# =============================================================================
print("\n" + "="*80)
print("[8/10] Training minimal model...")
print("="*80)

minimal_indices = [all_features.index(f) for f in minimal_features]
X_train_minimal = X_train_scaled[:, minimal_indices]
X_val_minimal = X_val_scaled[:, minimal_indices]
X_test_minimal = X_test_scaled[:, minimal_indices]

model_minimal = xgb.XGBRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=6,
    random_state=RANDOM_SEED,
    n_jobs=-1,
    tree_method='hist'
)

model_minimal.fit(X_train_minimal, y_train, verbose=False)

y_pred_val_minimal = model_minimal.predict(X_val_minimal)
y_pred_test_minimal = model_minimal.predict(X_test_minimal)

r2_val_minimal = r2_score(y_val, y_pred_val_minimal)
r2_test_minimal = r2_score(y_test, y_pred_test_minimal)
rmse_val_minimal = np.sqrt(mean_squared_error(y_val, y_pred_val_minimal))
rmse_test_minimal = np.sqrt(mean_squared_error(y_test, y_pred_test_minimal))

print(f"\n{'='*60}")
print(f"MINIMAL MODEL ({minimal_idx} features)")
print(f"{'='*60}")
print(f"  Val R¬≤:   {r2_val_minimal:.4f} | RMSE: {rmse_val_minimal:.4f}")
print(f"  Test R¬≤:  {r2_test_minimal:.4f} | RMSE: {rmse_test_minimal:.4f}")
print(f"  vs Full:  {(r2_val_minimal/r2_val_full*100):.1f}% performance")
print(f"{'='*60}")

# Save models
joblib.dump({
    'model_full': model_full,
    'model_minimal': model_minimal,
    'scaler': scaler,
    'encoders': encoders,
    'all_features': all_features,
    'minimal_features': minimal_features,
    'district_col': district_col,
    'crop_col': crop_col,
    'season_col': season_col,
    'yield_col': yield_col
}, os.path.join(OUTPUT_DIR, 'trained_models.pkl'))

print("\n‚úì Models saved")

# =============================================================================
# STEP 9: PER-CROP ANALYSIS
# =============================================================================
print("\n" + "="*80)
print("[9/10] Analyzing per-crop performance...")
print("="*80)

crop_results = []
for crop in df[crop_col].unique()[:20]:  # Top 20 crops
    crop_mask_test = df.loc[X_test.index, crop_col] == crop

    if crop_mask_test.sum() < 10:
        continue

    y_test_crop = y_test[crop_mask_test]
    y_pred_crop_full = y_pred_test_full[crop_mask_test]
    y_pred_crop_min = y_pred_test_minimal[crop_mask_test]

    r2_full = r2_score(y_test_crop, y_pred_crop_full)
    r2_min = r2_score(y_test_crop, y_pred_crop_min)

    crop_results.append({
        'crop': crop,
        'n_samples': crop_mask_test.sum(),
        'r2_full': r2_full,
        'r2_minimal': r2_min,
        'performance_retained': (r2_min/r2_full*100) if r2_full > 0 else 0
    })

crop_df = pd.DataFrame(crop_results).sort_values('r2_full', ascending=False)
print(f"\nTop Crops by Performance:")
print(crop_df.head(10).to_string(index=False))

# =============================================================================
# STEP 10: VISUALIZE & EXPORT
# =============================================================================
print("\n" + "="*80)
print("[10/10] Generating results...")
print("="*80)

# Create visualization
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
plt.ioff()

# 1. Predictions
ax = axes[0, 0]
ax.scatter(y_test, y_pred_test_full, alpha=0.5, s=10, label='Full')
ax.scatter(y_test, y_pred_test_minimal, alpha=0.5, s=10, label='Minimal')
ax.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
ax.set_xlabel('Actual Yield')
ax.set_ylabel('Predicted Yield')
ax.set_title(f'Predictions (Full R¬≤={r2_test_full:.3f}, Minimal R¬≤={r2_test_minimal:.3f})')
ax.legend()
ax.grid(alpha=0.3)

# 2. Feature Importance
ax = axes[0, 1]
top_n = min(20, len(importance_df))
top_imp = importance_df.head(top_n)
colors = ['coral' if f in minimal_features else 'steelblue' for f in top_imp['feature']]
ax.barh(range(top_n), top_imp['importance'].values, color=colors, alpha=0.8)
ax.set_yticks(range(top_n))
ax.set_yticklabels(top_imp['feature'].values, fontsize=7)
ax.invert_yaxis()
ax.set_xlabel('Importance')
ax.set_title('Top 20 Features (Orange=Minimal Set)')
ax.grid(axis='x', alpha=0.3)

# 3. Ablation
ax = axes[1, 0]
ax.plot(ablation_df['n_features'], ablation_df['r2'], marker='o', linewidth=2)
ax.axhline(y=r2_val_full, color='blue', linestyle='--', label='Full')
ax.axhline(y=target_performance, color='red', linestyle='--', label='95% target')
ax.axvline(x=minimal_idx, color='orange', linestyle='--', label=f'Minimal ({minimal_idx})')
ax.set_xlabel('Number of Features')
ax.set_ylabel('R¬≤ Score')
ax.set_title('Ablation Study')
ax.legend()
ax.grid(alpha=0.3)

# 4. Per-Crop
ax = axes[1, 1]
if len(crop_df) > 0:
    top_crops = crop_df.head(10)
    x_pos = np.arange(len(top_crops))
    width = 0.35
    ax.barh(x_pos - width/2, top_crops['r2_full'].values, width, label='Full', color='steelblue')
    ax.barh(x_pos + width/2, top_crops['r2_minimal'].values, width, label='Minimal', color='coral')
    ax.set_yticks(x_pos)
    ax.set_yticklabels(top_crops['crop'].values, fontsize=8)
    ax.invert_yaxis()
    ax.set_xlabel('R¬≤ Score')
    ax.set_title('Per-Crop Performance')
    ax.legend()
    ax.grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, 'results.png'), dpi=150)
print("‚úì Saved results.png")
plt.show()

# Save CSVs
importance_df.to_csv(os.path.join(OUTPUT_DIR, 'feature_importance.csv'), index=False)
ablation_df.to_csv(os.path.join(OUTPUT_DIR, 'ablation_results.csv'), index=False)
crop_df.to_csv(os.path.join(OUTPUT_DIR, 'per_crop_analysis.csv'), index=False)

minimal_features_df = pd.DataFrame({
    'rank': range(1, len(minimal_features) + 1),
    'feature': minimal_features,
    'importance': [importance_df[importance_df['feature']==f]['importance'].values[0]
                   for f in minimal_features]
})
minimal_features_df.to_csv(os.path.join(OUTPUT_DIR, 'minimal_features.csv'), index=False)

# =============================================================================
# SUMMARY
# =============================================================================
print("\n" + "="*80)
print("‚úÖ ANALYSIS COMPLETE!")
print("="*80)

print(f"\nüìä RESULTS:")
print(f"  Full Model:    {len(all_features)} features ‚Üí R¬≤ = {r2_test_full:.4f}")
print(f"  Minimal Model: {minimal_idx} features ‚Üí R¬≤ = {r2_test_minimal:.4f}")
print(f"  Efficiency:    {100-(minimal_idx/len(all_features)*100):.1f}% fewer features")
print(f"  Performance:   {(r2_test_minimal/r2_test_full*100):.1f}% retained")

print(f"\nüéØ MINIMAL FEATURES (Top {min(15, len(minimal_features))}):")
for i, feat in enumerate(minimal_features[:15], 1):
    print(f"  {i:2d}. {feat}")

print(f"\nüìÅ Output Files:")
print(f"  - trained_models.pkl")
print(f"  - minimal_features.csv")
print(f"  - feature_importance.csv")
print(f"  - ablation_results.csv")
print(f"  - per_crop_analysis.csv")
print(f"  - results.png")

# Download
print("\nüì• Creating download package...")
import shutil
shutil.make_archive('/content/results', 'zip', OUTPUT_DIR)
files.download('/content/results.zip')

print("\n‚úÖ Done! Check your downloads folder for results.zip")
print("="*80)
```
